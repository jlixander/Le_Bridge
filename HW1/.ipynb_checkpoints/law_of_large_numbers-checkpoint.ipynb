```python
import matplotlib.pyplot as plt
import random
from fractions import Fraction

num_trials = [100,1000,10000,100000,500000]
```

```python
def get_roll():
    return random.randint(1,6)
```

```python
def Average(a):
    return sum(a) / len(a)    
```

```python
print("Expected value: 3.5")
for tot_trial in num_trials:
    trial = 0
    expectation = []
    #print(tot_trial)
    
    while trial < tot_trial:
        expectation.append(get_roll())
        expectation_avg = Average(expectation)
        trial += 1
    print('Trials: {0} Trial average: {1}'.format(tot_trial, expectation_avg))
```
