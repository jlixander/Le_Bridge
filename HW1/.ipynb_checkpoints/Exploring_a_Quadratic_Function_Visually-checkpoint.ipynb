```python
from pylab import show, legend, title, xlabel, ylabel, axis
import matplotlib.pyplot as plt
```

```python
'''
Quadratic function calculator
'''
# Assume 10 values of x
x_values = [-1, 0, 1, 2, 3, 4, 5, 6, 7, 8]

#Create empty list to hold generated y-values
y_values = []

#Find y values based on assumed x values
for x in x_values:
    # Calculate the value of the quadratic function
    y = x**2 + 2*x + 1
    y_values.append(y)
    print('x={0} y={1}'.format(x, y))
    
print(y_values)
```

```python
plt.plot(x_values, y_values)
#legend([2000,2006,2012])
title('Quadratic Equation')
xlabel('x-axis')
ylabel('y-axis')
```
