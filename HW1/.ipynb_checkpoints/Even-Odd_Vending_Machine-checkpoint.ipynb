```python
#Get input from user and convert to int
VM_input = input("Please input a number")
VM_input = int(VM_input)

# #Identify parity of number
parity = VM_input % 2
#print(parity)

if parity == 1:
    print("The number is odd")
else:
    print("The number is even")
    
for i in range(VM_input-2, VM_input+18,2):
    print(i+2)
    
```
