Implement a statistics calculator that takes a list of numbers in the file
mydata.txt and then calculates and prints their mean, median, mode,
variance, and standard deviation using the functions we wrote earlier in
this chapter.

```python
#Libraries
from collections import Counter
```

```python
#Function to read .txt file
def read_data(filename):
    numbers = []
    with open(filename) as f:
        for line in f:
            numbers.append(float(line))
    return numbers

#Function to calc mean
def calculate_mean(numbers):
    s = sum(numbers)
    N = len(numbers)
    mean = s/N
    return mean

#Function to calc median
def calculate_median(numbers):
    N = len(numbers)
    numbers.sort()
    
    # Find the median
    if N % 2 == 0:
        # if N is even
        m1 = N/2
        m2 = (N/2) + 1
        # Convert to integer, match position
        m1 = int(m1) - 1
        m2 = int(m2) - 1
        median = (numbers[m1] + numbers[m2])/2
    
    else:
        m = (N+1)/2
        # Convert to integer, match position
        m = int(m) - 1
        median = numbers[m]
    return median

#Function to calc mode
def calculate_mode(numbers):
    c = Counter(numbers)
    numbers_freq = c.most_common()
    max_count = numbers_freq[0][1]
    modes = []
    
    for num in numbers_freq:
        if num[1] == max_count:
            modes.append(num[0])
    return modes

#Function to calc variance
def find_differences(numbers):
    # Find the mean
    mean = calculate_mean(numbers)
    
    # Find the differences from the mean
    diff = []
    
    for num in numbers:
        diff.append(num-mean)
    return diff

#Calc variance and std
def calculate_variance_std(numbers):
    # Find the list of differences
    diff = find_differences(numbers)
    # Find the squared differences
    squared_diff = []
    
    for d in diff:
        squared_diff.append(d**2)
    
    # Find the variance
    sum_squared_diff = sum(squared_diff)
    variance = sum_squared_diff/len(numbers)
    std = variance**0.5
    return variance, std



#Create a function that compiles all the stats funcs
def stats_calc(data):
    mean = calculate_mean(data)
    median = calculate_median(data)
    mode = calculate_mode(data)
    variance, std = calculate_variance_std(data)
    
    
    return mean, median, mode, variance, std
```

```python
if __name__ == '__main__':
    data = read_data('mydata.txt')
    mean, median, mode, variance, sd = stats_calc(data)
    print("The mean is: {0}".format(mean))
    print("The median is: {0}".format(median))
    print("The mode is: {0}".format(mode))
    print("The variance is: {0}".format(variance))
    print("The standard deviation is: {0}".format(sd))
```
