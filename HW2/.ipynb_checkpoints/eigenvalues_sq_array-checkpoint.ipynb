```python

def getEigenVec(A, values):
    x,y = symbols('x, y')
    
    eq1 = sp.Eq((A[0][0])*x + (A[0][1])*y , values[0])             # ax + by  = Lx
    eq2 = sp.Eq((A[0][1])*x + (A[1][1])*y, values[0])              # cx + dy  = Ly
    ans = sp.solve((eq1, eq2), (x, y))
    return ans
```

### Question 2

```python
#A = U · Σ · V T
# A is the input matrix
# U are the left singular vectors,
# sigma are the diagonal/eigenvalues
# V are the right singular vectors.

def fac_arr_svd(A):
    #Transpose matrix
    print(A)
    print("")
    
    AT = A.T
    print(AT)
    
    ATA = AT @ A
    print("")
    print(ATA)  
    
    #calc eigenvalues & vectors of ATA
    values, vectors = eig(ATA)
    print("")
    print(f"evals: {values}")
    
    print("")
    print(f"evecs aka V: {vectors}")
        
    #Calc transpose of ATA vectors
    VT = vectors.T
    print("")
    print(f"evecs aka V^T: {VT}")
    
    #Get singular values and place in matrix (desc order)
    sval1 = math.sqrt(values[0])
    sval2 = math.sqrt(values[1])
    S = np.zeros([n,n])
    if values[0]> values[1]:
        S[0][0] = sval1
        S[1][1] = sval2
    else:
        S[0][0] = sval2
        S[1][1] = sval1
    print("")
    print(f"S: {S}")
        
    #Calc inverse of S
    S_inv = inv(S)
    print("")
    print(f"S^-1: {S_inv}")
    
    
    #ATA-10I and get eigenvectors
    ATA1 = ATA
    ATA1[0][0] = ATA1[0][0]-10
    ATA1[1][1] = ATA1[1][1]-10
    print("")
    print(ATA1)
    
    #ATA-40I and get eigenvectors
    ATA2 = ATA
    ATA2[0][0] = ATA2[0][0]-40
    ATA2[1][1] = ATA2[1][1]-40
    print("")
    print(ATA2)
    
    #Get U (U=AVS^-1)
    U = A.dot(vectors).dot(S_inv)
    print("")
    print(U)
    
    #final check
    A = U.dot(S).dot(VT)
    print("")
    print(f'Confirmation: {A}')

    
    
    
```

```python
fac_arr_svd(A)
```

```python
test_arr = array([
    [2, -5, 3],
    [0, 7, -2],
    [-1, 4, 1]])
```

```python
#Get the determinant of an array
def get_det (A):
    if len(A) == 2:
        det = (A[0][0]*A[1][1])-(A[0][1]*A[1][0])
    if len(A) == 3:
        det = A[0][0]*(A[1][1]*A[2][2]-A[1][2]*A[2][1])-A[0][1]*(A[1][0]*A[2][2]-A[1][2]*A[0][2])+A[0][2]*(A[1][0]*A[2][1]-A[0][2]*A[1][1])
    return det
```

```python
print(get_det(test_arr))
```
